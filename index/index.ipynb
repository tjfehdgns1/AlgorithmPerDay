```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
```

```python
plt.style.use(['seaborn-v0_8-notebook'])
```

## Line Plot

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure()
plt.plot([0, 0.2, 0.4, 0.6, 0.8, 1] * 5)
```

```python
x = np.arange(0, 10, 0.01)
fig = plt.figure()
plt.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

## Line Style

||
|---|---|
'-'|'soild'
'--'|'dashed'
'-.'|'dashdot'
':'|'dotted'

```python

plt.plot(x, np.sin(x), linestyle='-.')
plt.plot(x, np.cos(x), linestyle='--')
```

## Color Style

```python
plt.plot(x, np.sin(x), color='g')
plt.plot(x, np.cos(x), color='r')
```

```python
plt.plot(x, np.sin(x), 'g--')
plt.plot(x, np.cos(x), 'c-')
```

## Plot Axis

```python
plt.plot(np.random.randn(50))
plt.xlim(-1, 50)
plt.ylim(-5, 5)
```

```python
plt.plot(np.random.randn(50))
plt.axis([-1, 50, -5, 5])
```

```python
plt.plot(np.random.randn(50))
plt.axis('tight')
```

## Plot Label

```python
plt.plot(np.random.randn(50), label='A')
plt.plot(np.random.randn(50), label='B')
plt.title('Practice')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
```

## Font Manager

```python
set([f.name for f in mpl.font_manager.fontManager.ttflist])
```

```python
font1 = {'family': 'DejaVu Sans', 'size': 24, 'color': 'black'}
plt.plot(np.random.randn(50), label='A')
plt.title('title', fontdict=font1)
```

## Plot Legend
|||
|---|---|
0|best|
1|upper right|
2|upper left
3|lower left
4|lower right
5|right
6|center left
7|center right
8|lower center
9|upper center

```python
fig, ax = plt.subplots()
ax.plot(np.random.randn(10), 'r-', label='A')
ax.plot(np.random.randn(10), 'g--', label='B')
ax.plot(np.random.randn(10), 'b-.', label='C')
ax.axis('equal')
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(np.random.randn(10), 'r-', label='A')
ax.plot(np.random.randn(10), 'g--', label='B')
ax.plot(np.random.randn(10), 'b-.', label='C')
ax.axis('equal')
ax.legend(loc='upper center', frameon=False, ncol=2)
```

```python
plt.figure(figsize=(8, 4))
x = np.linspace(0, 10, 1000)
y = np.cos(x[:, np.newaxis] * np.arange(0, 2, 0.4))
lines = plt.plot(x, y)
plt.legend(lines[:3], ['c1', 'c2', 'c3'])
```
